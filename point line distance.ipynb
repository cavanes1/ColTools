# Distance from point to line calculator

```python
# https://stackoverflow.com/questions/39840030/distance-between-point-and-a-line-from-two-points
def pld(P1, P2, P3):
    # convert points to numpy arrays
    P1 = np.array(P1)
    P2 = np.array(P2)
    P3 = np.array(P3)
    
    # print P4
    P4 = P1 + (np.dot(P3 - P1, P2 - P1) / (np.linalg.norm(P2 - P1))**2)*(P2 - P1)
    print("\nPoint on line closest to CI:")
    for i in range(len(P4)):
        print(str(i + 1) + "   " + str(P4[i]))

    # difference
    dff = P3 - P4
    print("\nVector between point and line:")
    for i in range(len(dff)):
        print(str(i + 1) + "   " + str(dff[i]))

    # My adjusted distance
    print("\nDistance between point and line:")
    print(np.linalg.norm(dff))
    print()

    # Stackexchange exact answer for distance
    #print("\nStackexchange's distance:")
    #print(np.linalg.norm(P3 - np.dot(P3 - P1, P2 - P1) / (np.linalg.norm(P2 - P1))**2))
```
